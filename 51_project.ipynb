```python
import pandas as pd

df = pd.read_csv("flights_sample_3m.csv")

# Replace NaN in numerical columns with mean
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

print("NaN values replaced with mean")
print(df.isnull().sum())
```

```python
import pandas as pd

df = pd.read_csv("flights_sample_3m.csv")

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

print("NaN values replaced with median")
```

```python
import numpy as np

# replace missing 'airline' values by most frequent
df["AIRLINE"] = df["AIRLINE"].replace(np.nan, "UA")

# simply drop whole row with NaN in "arr_delay" column
df.dropna(subset=["ARR_DELAY"], axis=0, inplace=True)

# reset index, because we dropped rows
df.reset_index(drop=True, inplace=True)
```

```python


# Load dataset
df = pd.read_csv("flights_sample_3m.csv")

# Select numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply standardization (Z-score)
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()

print("Data Standardized Successfully")
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Select numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply scaling
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Standardization using sklearn done")
```

```python


# Load dataset
df = pd.read_csv("flights_sample_3m.csv")

# Select numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply Min-Max Normalization
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())

print("Data Normalized Successfully")
print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Select numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply normalization
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Normalization using sklearn done")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("flights_sample_3m.csv")

# Create bins for departure delay
bins = [-100, 0, 30, 60, 1000]   # ranges
labels = ["Early/On-time", "Low Delay", "Medium Delay", "High Delay"]

df["dep_delay_binned"] = pd.cut(df["DEP_DELAY"], bins=bins, labels=labels)

print(df[["DEP_DELAY", "dep_delay_binned"]].head())
```
